```python
df.plot(kind='scatter',
       x='Height',
       y='Weight',
       title='Weight and Height in adults')

plt.plot([55,78], [75,250], color='purple', linewidth=4)
```

```python
def line(x, w=0, b=0):
    return x * w + b

x = np.linespace(55, 80, 100)
yhat = line(x, w=0, b=0)

df.plot(kind='scatter',
       x='Height',
       y='Weight',
       title='Weight and Height in adults')
plt.plot(x, yhat, color='orange', linewidth=3)
```
