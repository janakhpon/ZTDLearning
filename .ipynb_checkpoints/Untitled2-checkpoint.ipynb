```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from pandas import DataFrame as df
```

```python
df = pd.read_csv('../data/weight-height.csv')
```

```python
df['Gender'].value_counts()
```

```python
_=df.plot(kind='scatter', x="Weight", y="Height")
```

```python
males = df[df['Gender'] == 'Male']
females = df.query('Gender == "Female"')
fig, ax = plt.subplots()

males.plot(kind='scatter', x='Height', y='Weight', ax=ax, color='blue', alpha=0.3, title='Male & Female Populations')

females.plot(kind='scatter', x='Height', y='Weight', ax=ax, color='red', alpha=0.3)
```
